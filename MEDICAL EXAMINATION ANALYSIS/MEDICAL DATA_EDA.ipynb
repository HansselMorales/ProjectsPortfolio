```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading file

```python
df = pd.read_csv('medical_examination.csv')
df.describe()
```

## Adding a new column named "overweight"

#### Determining if a person is overweight

```python
#Calculating their BMI by dividing their weight in kilograms by the square of their height in meters
BMI = df['weight'] / (df['height']/100)**2
df['overweight'] = BMI
df
```

```python
#If BMI value is > 25 then the person is overweight (value = 1 for overweight)
df.loc[df['overweight']>25, 'overweight'] = 1
df.head(30)
```

```python
#Value 0 for NOT overweight
df.loc[(df['overweight']<=25) & (df['overweight']>1), 'overweight'] = 0
df
```

## Normalizing data

#### 0 is good, 1 is bad

```python
#If the value of cholesterol or gluc is 1, value = 0. If the value is more than 1, value = 1
df.loc[df['cholesterol'] == 1, 'cholesterol'] = 0
df.loc[df['cholesterol'] > 1, 'cholesterol'] = 1
df.loc[df['gluc'] == 1, 'gluc'] = 0
df.loc[df['gluc'] > 1, 'gluc'] = 1
df
```

## Seaborn's catplot( ) chart

#### Converting data into long format

```python
value_list = ['active','alco','cholesterol','gluc','overweight','smoke']
df_long = pd.melt(df, id_vars='cardio', value_vars=value_list, var_name='variable', value_name='value')
df_long
```

#### Drawing the catplot( ) chart

```python
sns.catplot(data=df_long, x= 'variable', kind='count', hue='value', col='cardio', height=5, aspect=1.5)
```

## Cleaning data

```python
df
```

#### Filtering out patient segments: diastolic pressure is higher than systolic (Keeping the correct data with (df['ap_lo'] <= df['ap_hi']))

```python
df.loc[df['ap_lo']>df['ap_hi']]
```

```python
#Creating new columns called 'number_index' in order to copy 'id' column's values
index_list = []
for i in range(len(df['id'])):
    index_list.append(i)
index_list
df['number_index'] = index_list
df
```

```python
# Creating a list with incorrect pressure values
incorrect_pressure = df.loc[df['ap_lo']>df['ap_hi'],'number_index']
pressure_list = []
for i in incorrect_pressure:
    pressure_list.append(i)

pressure_list
```

```python
df.loc[df['ap_lo']>df['ap_hi'],'number_index']
```

```python
# Dropping incorrect pressure values
df.drop(pressure_list,axis=0, inplace=True)
```

```python
df.loc[df['ap_hi']>=df['ap_lo']]
```

#### Filtering out patient segments: height is less than the 2.5th percentile and more than the 97.5th percentile (Keeping the correct data with (df['height'] >= df['height'].quantile(0.025)) and (df['height'] <= df['height'].quantile(0.975)))

```python
df
```

```python
df.loc[df['height']<df['height'].quantile(0.025)]
```

```python
df.loc[df['height']>df['height'].quantile(0.975)]
```

```python
# Creating a list with incorrect height values
height_list= []
incorrect_height = df.loc[(df['height']<df['height'].quantile(0.025)) | (df['height']>df['height'].quantile(0.975)), 'number_index']

for i in incorrect_height:
    height_list.append(i)

height_list
```

```python
# Dropping incorrect height values
df.drop(height_list,axis=0,inplace=True)
df
```

#### Filtering out patient segments: weight is less than the 2.5th percentile and more than the 97.5th percentile (Keeping the correct data with (df['weight'] >= df['weight'].quantile(0.025)) and (df['weight'] <= df['weight'].quantile(0.975)))

```python
df.loc[df['weight']<df['weight'].quantile(0.025)]
```

```python
df.loc[df['weight']>df['weight'].quantile(0.975)]
```

```python
# Creating a list with incorrect weight values
weight_list = []
incorrect_weight = df.loc[(df['weight']<df['weight'].quantile(0.025)) | (df['weight']>df['weight'].quantile(0.975)), 'number_index']

for i in incorrect_weight:
    weight_list.append(i)

weight_list
```

```python
df.drop(weight_list,axis=0,inplace=True)
df
```

```python
# Dropping incorrect weight values
df.drop('number_index',axis=1, inplace=True)
df
```
