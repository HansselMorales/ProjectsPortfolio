```python
import pandas as pd
import json
```

```python
with open ('t20_json_files/t20_wc_match_summary.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head(30)
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard' :'match_id'}, axis=1, inplace=True)
df_match.head()
```

```python
df_match.loc[df_match.isnull()['margin'] == True]
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
```

**Batting Summary**

```python
with open ('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head(10)
```

```python
df_batting.info()
```

```python
df_batting.loc[df_batting['dismissal'].isnull() == True]
```

```python
df_batting['dismissal'].fillna('', inplace=True)
df_batting
```

```python
df_batting['out/not_out'] = df_batting['dismissal'].apply(lambda x: 'out' if len(x) > 0 else 'not_out')
df_batting.head()
```

```python
df_batting.drop('dismissal', axis=1, inplace=True)
df_batting.head(11)
```

**Cleaning up weird characters**

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv', index = False)
```

**Bowling Summary**

```python
with open ('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    
    all_record_bowling = []
    for i in data:
        all_record_bowling.extend(i['bowlingSummary'])

df_bowling = pd.DataFrame(all_record_bowling)
df_bowling
```

```python
df_bowling.info()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head(30)
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```

**Players Information**

```python
with open ('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
    
df_players = pd.DataFrame(data[3:])
df_players.head()
```

```python
df_players.drop('input', axis=1, inplace=True)
df_players
```

```python
df_players.loc[df_players['team'] == 'India']
```

```python
df_players['bowlingStyle'].fillna('', inplace=True)
df_players['description'].fillna('', inplace=True)
df_players.head()
```

**Cleaning up weird characters**

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head()
```

```python
df_players.to_csv('t20_csv_files/dim_players_no_images.csv', index = False)
```
